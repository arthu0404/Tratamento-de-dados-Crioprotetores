# **Extraindo Tabela de Calibração do PDF**

```python
import pdfplumber
import csv

pdf_path = r"..\calib_temp\cryojet_24-02-2025_11-30-14_renata_experiment.pdf"
csv_path = "../extracted/tabela_extraida_calib_temp.csv"

with pdfplumber.open(pdf_path) as pdf, open(csv_path, mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            writer.writerows(table)
```
